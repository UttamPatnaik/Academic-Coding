## Aim: Implement assymetric cryptography using RSA algorithm and Diffie-Hellman Key Exchange to establish a shared secret key between communicating parties.

```python
import random
# Key Generation
def gcd(m,n):
    if(n==0):
        return m
    else:
        return gcd(n,m%n)
def key_gen(p, q):
    n=p*q
    phin=(p-1)*(q-1)
    e=private_key(phin-1)
    d=public_key(e,phin-1)
    return e,d
def private_key(t):
    e=random.randint(2,t)
    if(gcd(e,t)!=1):
        e=random.randint(2,t)
        return gcd(e,t)
    return e
def public_key(et,p):
    d=random.randint(2,p)
    if((d*et)%p==1):
        return d
    else:
        return public_key(et,p)
x=int(input("Enter a prime number: "))
y=int(input("Enter another prime number: "))
a,b=key_gen(x,y)
print(a)
print(b)
```
