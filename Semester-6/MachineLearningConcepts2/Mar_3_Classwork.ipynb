```python
import numpy as np


class Perceptron:
    def __init__(self, lr=0.01, epochs=10):
        self.lr = lr
        self.epochs = epochs

    def activation(self, x):
        return np.where(x >= 0, 1, -1)

    def fit(self, X, y):
        # Initialize weights and bias
        self.weights = np.zeros(X.shape[1])
        self.bias = 0

        # Training loop
        for _ in range(self.epochs):
            for i in range(len(X)):
                linear = np.dot(X[i], self.weights) + self.bias
                y_pred = self.activation(linear)

                update = self.lr * (y[i] - y_pred)

                # Update weights and bias
                self.weights += update * X[i]
                self.bias += update

    def predict(self, X):
        X = np.atleast_2d(X)  # ensure 2D input
        linear = np.dot(X, self.weights) + self.bias
        return self.activation(linear)


# ----------------------------
# Example Usage
# ----------------------------

# Linearly separable dataset
X = np.array([
    [2, 3],
    [1, 1],
    [-1, -1],
    [-2, -3]
])

y = np.array([1, 1, -1, -1])

# Create and train model
p = Perceptron(lr=0.1, epochs=10)
p.fit(X, y)

# Test prediction
test_sample = np.array([[2, 2]])
prediction = p.predict(test_sample)

print("Prediction:", prediction)
```
