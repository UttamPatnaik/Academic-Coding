# Reinforcement Learning

## What is Reinforcement Learning?

Reinforcement Learning is a type of **Machine Learning (ML)** in which an agent learns how to behave in an environment by performing actions and receiving feedback in the form of **rewards** and **penalties**.

Unlike supervised learning, there is no labeled dataset. Instead, the system learns through **trial-and-error** interactions.

---

## Trial and Error Learning

Trial-and-error learning involves two important strategies:

* **Exploration**
* **Exploitation**

### Exploitation
When the agent takes known steps that have previously proven successful in yielding high reward values, it is called **exploitation**.

### Exploration
When the agent tries out unknown paths and, by doing so, identifies new ways to achieve higher rewards, it is called **exploration**.

---

## Goal
**Maximize cumulative reward over time.**

## Examples
* A robot learning to walk
* A computer learning to play chess or Go
* Self-driving cars learning navigation

---

## Elements of Reinforcement Learning

### Agent
The learner or decision-maker that takes actions in the environment.

### Environment
The external system with which the agent interacts. It responds to the agent's actions and provides feedback in the form of rewards or penalties.

Based on the properties of the environment, it can be classified as:

* **Deterministic vs Stochastic** 
    * **Deterministic:** Outcome is based on current state and action (Example: New state of a piece in chess when moved).  
    * **Stochastic:** Outcome is not certain and only a probability of possible outcomes can be estimated (Example: Outcome of dice is not predictable).

* **Fully Observable vs Partially Observable** 
    * **Fully Observable:** When the system state is visible to the agent at all times (Example: All players can see the position of all pieces on the board in chess).  
    * **Partially Observable:** If the agent can't determine the state of the whole system all the time (Example: In poker, players can't see the cards in other players' hands).

* **Discrete vs Continuous** 
    * **Discrete:** Finite set of actions is possible for moving from one state to another (Example: Number of possible actions in chess is limited).  
    * **Continuous:** Set of actions available is infinite (Example: Infinite possible routes of travelling from one place to another).

### State
Describes the current situation of the agent within the environment.

**Examples:**
* Current position of a robot
* Board configuration in a game
* Current traffic situation for a vehicle

### Action
It is a decision taken by the agent that influences the environment according to its strategy or policy. The collection of all possible actions is called the **action space**.

### Policy
It is a strategy that the agent uses to achieve the goal.

### Reward
It is a numerical value given to the agent after it performs an action. It indicates whether the action was good/bad; the effect was positive or negative, which is equivalent to reward and punishment.

### Value Function
It defines long-term gain while reward defines immediate gain. When an agent takes an action in the current state, it can expect a certain reward. The action also affects the subsequent state's actions.

### Model
It is a replica of the environment that imitates the behaviour of the environment. It helps in deriving inferences about future courses of action based on prior knowledge about how the environment will behave in such a situation.

The technique of solving reinforcement learning problems using models is called the **model-based approach**, and the main aim of the agent is to explore the environment to learn the model.

The techniques that depend on **exploration and exploitation** without relying on a predefined model of the environment are termed the **model-free approach**.

## Reinforcement Learning Principles

---

### Markov Property

It is a fundamental concept in reinforcement learning which states that the **future is independent of the past, given the present.** In other words, the current state $s_t$ captures all relevant information from the history of the process, so knowing the previous states is unnecessary for predicting the next state $s_{t+1}$.



**Formula:**
A state $S_t$ is Markov if and only if:
$$P[S_{t+1} | S_t] = P[S_{t+1} | S_1, ..., S_t]$$

### Transition Probability

It is the probability that the agent will move from one state $s$ to a successor state $s'$ after taking an action. For a Markov state $s$ and successor state $s'$, the state transition probability is defined by:

**Formula (Matrix):**
For a finite set of states, the transition probability can be represented as a **State Transition Matrix** $\mathcal{P}$, where each row sums to 1:

$$\mathcal{P} = \begin{bmatrix} 
P_{11} & P_{12} & \dots & P_{1n} \\
P_{21} & P_{22} & \dots & P_{2n} \\
\vdots & \vdots & \dots & \vdots \\
P_{n1} & P_{n2} & \dots & P_{nn}
\end{bmatrix}$$

Where $P_{ij}$ is the probability of transitioning from state $i$ to state $j$.

### Markov Process or Markov Chain

It is a stochastic process that satisfies the Markov property; i.e., it is a sequence of random states $S_1, S_2, ...$ that satisfies the Markov property. It is essentially a memoryless random process defined by a tuple $\langle \mathcal{S}, \mathcal{P} \rangle$.

* **$\mathcal{S}$**: A finite set of states.
* **$\mathcal{P}$**: A state transition probability matrix.



---

### Markov Reward Process (MRP)

It is an extension of a Markov process that includes rewards associated with states. It allows us to evaluate how good a state is based on expected rewards obtained from it. It is defined by a tuple $\langle \mathcal{S}, \mathcal{P}, \mathcal{R}, \gamma \rangle$:

* **$\mathcal{S}$**: A finite set of states.
* **$\mathcal{P}$**: A state transition probability matrix.
* **$\mathcal{R}$**: A reward function, where $R_s = E[R_{t+1} | S_t = s]$.
* **$\gamma$**: A discount factor, $\gamma \in [0, 1]$.

Thus an MRP describes how rewards accumulate as the system transitions between states.

### Reward
Numerical values an agent receives based on the action taken while being in a state in the environment.

---

### Bellman Equation for Value Function

It provides a recursive relationship for the value function.

$$v_{\pi}(s) = \sum_{a \in A} \pi(a|s) \sum_{s', r} p(s', r | s, a) [r + \gamma v_{\pi}(s')]$$

Where,
* **$v_{\pi}(s)$: value of state sss under policy $\pi$
* **$\pi(a|s)$: The probability of taking action $a$ in state $s$ under policy $\pi$.
* **$p(s', r | s, a)$: The probability of transitioning to state $s'$ with reward $r$, given action $a$ was taken in state $s$.
* **$\gamma$**: discount factor$.

### State-Value Function for a Policy $\pi$

The state-value function $v_{\pi}(s)$ represents the expected return starting from state $s$ and following policy $\pi$. It is defined as:

$$v_{\pi}(s) = \mathbb{E}_{\pi} [G_t | S_t = s]$$

Where:
* **$\mathbb{E}_{\pi}$**: The expected value if the agent follows policy $\pi$.
* **$G_t$**: The cumulative discounted return.
* **$S_t = s$**: The state at time step $t$ is $s$.

---

### Finite Markov Decision Process

It is a mathematical model where the agent makes some decision or takes action based on its cureent statte following a policy.

It is a tuple of $(S, A, P_{xy}', R(s|a), \gamma)$ where:

* **S** is the set of states
* **A** is the set of actions that agent can take
* **$P_{xy}'$** is the transition probability matrix
* **R(s|a)** is the reward function
* **$\gamma$** is the discount function
