## 1Ô∏è‚É£ Introduction

Hierarchical Clustering is an unsupervised learning technique used to group data into a **tree-like hierarchical structure** rather than forming a fixed number of clusters.

It is particularly useful when:

- The dataset contains **natural nested groupings**
- The number of clusters is **unknown**
- A **hierarchical relationship** between data points needs to be visualized

---

## 2Ô∏è‚É£ Applications

- üè¢ Organizing employees into departments and roles  
- üß¨ Gene classification in bioinformatics  
- üõç Customer segmentation  
- üìÇ Document clustering  
- üîé Discovering structure in unlabeled data  

---

## 3Ô∏è‚É£ Types of Hierarchical Clustering

### üîπ 3.1 Agglomerative Clustering (Bottom-Up)

- Each data point starts as its own cluster.
- The closest clusters are merged step by step.
- Process continues until all points form a single cluster.

‚úî Most commonly used approach.

---

### üîπ 3.2 Divisive Clustering (Top-Down)

- Start with all data points in one cluster.
- Recursively split clusters into smaller clusters.
- Continue until each point forms its own cluster.

‚úî Computationally more expensive than agglomerative.

---

## 4Ô∏è‚É£ Dendrogram

A **Dendrogram** is a tree-like diagram that visualizes the hierarchical merging or splitting process.

It helps to:
- Determine the optimal number of clusters  
- Understand hierarchical relationships  
- Visualize distance at which clusters merge  

### Example Representation

```
1  2  3  4  5   (Level 0)
 \/   |  |  |
 12   3  4  5   (Level 1)
 ||    \/   |  
 12    34   5   (Level 2)
 ||      \ /
 12      345    (Level 3)
   \    /
    \  /
     \/
    12345        (Level 4)
```

# 5Ô∏è‚É£ Distance Measures Between Clusters

Let:

- \( C_i \), \( C_j \) ‚Üí Two clusters  
- \( n_i \), \( n_j \) ‚Üí Number of data points  
- \( P_i \in C_i \), \( P_j \in C_j \) ‚Üí Data points  

---

## üîπ 5.1 Single Linkage (Nearest Neighbor)

\[
D_{\min}(C_i, C_j) = \min_{P_i \in C_i, P_j \in C_j} d(P_i, P_j)
\]

**Characteristics:**
- Uses minimum distance
- Sensitive to chaining effect
- Sensitive to noise and outliers

---

## üîπ 5.2 Complete Linkage (Farthest Neighbor)

\[
D_{\max}(C_i, C_j) = \max_{P_i \in C_i, P_j \in C_j} d(P_i, P_j)
\]

**Characteristics:**
- Uses maximum distance
- Produces compact clusters
- More robust than single linkage

---

## üîπ 5.3 Average Linkage

\[
D_{\text{avg}}(C_i, C_j) =
\frac{1}{n_i n_j}
\sum_{P_i \in C_i} \sum_{P_j \in C_j} d(P_i, P_j)
\]

**Characteristics:**
- Uses average pairwise distance
- More stable
- Less sensitive to outliers

# 9Ô∏è‚É£ Density-Based Methods ‚Äî DBSCAN

When using **partitioning methods (e.g., K-Means)** and **hierarchical clustering**, the resulting clusters are usually **spherical or convex in shape**.

However, these methods struggle to capture:

- Arbitrarily shaped clusters  
- S-shaped clusters  
- Uneven density clusters  
- Clusters separated by noise  

---

## üîé Why Density-Based Clustering?

The density-based clustering approach overcomes these limitations by:

- Identifying **dense regions** in the dataset  
- Separating them from **sparse regions (noise)**  
- Forming clusters based on **data point connectivity**

Instead of relying on distance to centroids, it focuses on **point density**.

---

## üìå DBSCAN  
**DBSCAN (Density-Based Spatial Clustering of Applications with Noise)** is one of the most popular density-based clustering algorithms.

---

### üîπ Key Idea

DBSCAN forms clusters by:

- Connecting points that lie in **high-density regions**
- Marking low-density points as **noise (outliers)**

---

## ‚úÖ Advantages

- Detects **arbitrarily shaped clusters**
- Automatically identifies **outliers**
- No need to specify number of clusters beforehand

---

## ‚ö†Ô∏è Limitations

- Sensitive to choice of Œµ and MinPts  
- Struggles with datasets having **varying densities**

---

# üîö Summary

Unlike partitioning and hierarchical clustering methods that tend to produce spherical clusters, **DBSCAN** effectively identifies complex-shaped clusters by analyzing data density.

This makes it especially useful for real-world datasets with irregular cluster structures.

# üî¨ Practical Implementation

## üå≥ Hierarchical Clustering  
### (Agglomerative Method ‚Äî Ward Linkage)

````python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
import numpy as np

X = np.random.randint(0, 10, size=(10, 2))

# Perform hierarchical clustering using Ward's method
# Ward's method minimizes the total within-cluster variance
Z = linkage(X, method='ward')

# Plot the dendrogram
plt.figure(figsize=(8, 5))
dendrogram(Z)

plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
````

# üî¨ Practical Implementation ‚Äî DBSCAN  
### (Density-Based Clustering)

````python
from sklearn.cluster import DBSCAN
dbscan=DBSCAN(eps=1.5, min_samples=2)
#eps: Radius of clusters
db_labels=dbscan.fit_predict(X)
plt.scatter(X[:,0],X[:,1],c=db_labels)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("DBSCAN Clustering")
plt.grid(True)
plt.show()
````
