## Regularization Techniques

Regularization is an extension made to another method (Regression)  
that penalizes models based on their complexity (overfitting),  
favouring simpler models that generalize better.

### Overfitting

- Training error << Test error  
- If both errors are about 80%, then the model is considered good  
- Happens when the model is too tuned towards the training data  

### Different Types of Regularization Techniques

- Ridge Regression  
- Lasso Regression (Least Absolute Shrinkage & Selection Operator)  
- Elastic Net (Ridge Regression + Lasso Regression)  
- Least Angle Regression  

### Ridge Regression

```python
import numpy as np
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import ElasticNet,LinearRegression
from sklearn.metrics import mean_squared_error

# 1. Generate dataset
X, y = make_regression(n_samples=200, n_features=5, noise=10, random_state=0)

# 2. Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# 3. Create model
model = ElasticNet(alpha=1.0, l1_ratio=0.5)

# 4. Train model
model.fit(X_train, y_train)

# 5. Predictions
y_pred = model.predict(X_test)

# 6. Evaluation
training_error=mean_squared_error(y_train, model.predict(X_train))
print("Training Error: ",training_error)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

lr=LinearRegression()
lr.fit(X_train,y_train)
y_predlr=lr.predict(X_test)
MSE_lr=mean_squared_error(y_train, model.predict(X_train))
print(MSE_lr)
```
