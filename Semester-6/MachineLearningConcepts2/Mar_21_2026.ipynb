### Instance-Based Learning (Memory-Based Learning)

Instance-based learning is a type of **machine learning** where the model learns from the **training instances** directly and **stores them** in memory. When a new instance (data point) is encountered, it is **compared** with the stored instances to make predictions.

### Key Characteristics:

* **No Explicit Generalization**: The model doesn’t explicitly generalize from the training data. Instead, it compares new data points to the stored instances.
* **Memory-Based**: The model keeps the training instances in memory and uses them for prediction during testing. The idea is to compare a new input to the training examples and output a prediction based on similarity.

### Common Instance-Based Learning Algorithms:

* **K-Nearest Neighbors (KNN)**: Classifies based on the majority class of the nearest neighbors.
* **Support Vector Machines (SVM)**: While technically not purely instance-based, SVM can be thought of as finding support vectors that define decision boundaries.
* **Radial Basis Function (RBF)**: A function used in neural networks and kernel methods, where the output depends on the distance from a center. This can be used in classification/regression tasks.
* **Principal Component Analysis (PCA)**: Not purely instance-based, but it involves transforming data into a new set of coordinates, which can be thought of as learning based on the instance relations.

### Radial Basis Function (RBF) Calculation

The **Radial Basis Function (RBF)** kernel is often used in **neural networks** and **support vector machines**. It calculates the similarity between a data point and a center (or reference point). The RBF is a **Gaussian** function that can be represented as:

[
\text{RBF}(x, c) = \exp \left( -\frac{| x - c |^2}{2 \sigma^2} \right)
]

Where:

* ( x ) is the input vector (a new instance).
* ( c ) is the center or reference point (could be a training instance or the mean of multiple instances).
* ( \sigma ) is a parameter that controls the width of the kernel, determining how sensitive the function is to the distance between ( x ) and ( c ).
* ( | x - c | ) is the **Euclidean distance** between the input vector ( x ) and center ( c ).

The RBF value decreases as the distance between ( x ) and ( c ) increases. If ( x ) is close to ( c ), the value will be high (closer to 1), and if ( x ) is far from ( c ), the value will be low (closer to 0).

### Using RBF for Prediction:

In an instance-based model, the RBF kernel is used to measure the similarity between new data points and stored instances. For classification tasks, the output can be the class with the highest similarity. For regression, it might involve weighted averaging of the target values corresponding to the nearest instances.

```python
import numpy as np

# Sample data
X = np.array([[1, 2], [2, 3], [3, 4], [4, 5]])  # Input features
Y = np.array([[0], [1], [1], [0]])  # Target labels (example)

# Define sigma (spread of the RBF kernel)
sigma = 1.0

# Radial Basis Function kernel
def rbf(x, c):
    return np.exp(-np.linalg.norm(x - c) ** 2 / (2 * sigma ** 2))

# Centers (could be the same as input X for simplicity in this example)
centers = X

# Construct the RBF kernel matrix H
H = np.array([[rbf(x, c) for c in centers] for x in X])

# Compute weights using the pseudoinverse
w = np.linalg.pinv(H) @ Y

# Make predictions
y_pred = H @ w

# Convert to binary class (threshold at 0.5)
y_pred_class = (y_pred > 0.5).astype(int)

# Output the predicted classes
print("Predicted: ", y_pred_class)
```
