# Artificial Neural Network (ANN)

## Architecture of ANN

Artificial Neural Networks can have different architectures depending on how neurons are organized and how signals flow between them.

### Types of ANN Architectures

- **Single Layer Feed Forward ANN**
- **Multi Layer Feed Forward ANN**
- **Competitive Network**
- **Recurrent Network**

## Learning Process of ANN

ANN learns by adjusting its structure and parameters during training.

### Key Factors in ANN Learning

- **Number of Layers**
  - Determines the depth of the network
- **Number of Hidden Layers**
  - Controls the complexity of the model
- **Number of Neurons in the Hidden Layer**
  - Affects the network’s ability to learn patterns
- **Signal Flow Direction**
  - **Feed Forward**
  - **Recurrent**
- **Weights**
  - Adjusted during training to minimize error

---

## Data Aspect and Suitable Network Type

| Data Aspect | Suitable Network |

## Practical Part

### Backpropagation on Single Layer

```python
import numpy as np

X=np.array([[0,0],[0,1],[1,0],[1,1]])
Y=np.array([[0],[1],[1],[0]])

np.random.seed(1)
weights=np.random.rand(2,1)
learning_rate=0.1

def sigmoid(x):
    return 1/(1+np.exp(-x))
    
def sigmoid_derivative(x):
    return x*(1-x)
    
for epoch in range(5000):
    output=sigmoid(np.dot(X,weights))
    error=Y-output
    adjustments=learning_rate*np.dot(X.T,error*sigmoid_derivative(output))
    weights+=adjustments

print("Final Output: ")
print(output)
```
