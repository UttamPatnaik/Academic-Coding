## 📖 3. Semi-Supervised Learning

### 📊 Notations
- **P** = Total Population  
- **P(K,i)** = Labelled data points at *i-th iteration*  
- **P(U,i)** = Unlabelled data points at *i-th iteration*  
- **P(C,i)** = Selected data from **P(U,i)**  

➡️ Condition:  
P(C,i) ⊆ P(U,i) (selected for labeling at iteration *i*)

---

### 🔍 Research Goal
To find the **best method** for identifying:  
P(C,i) ⊆ P(U,i)

---

### 📈 Example (P = 100 Data Points)

| Iteration (i) | Labelled (Train) | Unlabelled (Used for Prediction) |
|--------------|------------------|----------------------------------|
| 1            | 50               | 50                               |
| 2            | 60               | 40                               |
| 3            | 70               | 30                               |
| 4            | 80               | 20                               |
| 5            | 90               | 10                               |

---

## 🔁 Heuristic for Active Learning

1. Start with **P(U,i)** labelled=np.arange(10) 
2. Pick a subset **P(C,i)**  
3. Repeat the process

## 🤖 Support Vector Machine (SVM)

### 📌 Concepts

- **Minimal Margin Hyperplane Technique**  
  → Data with **minimum margin** should be selected for labeling  

- **Maximal Margin Hyperplane Technique**  
  → Data with **maximum margin separation** defines the boundary  

- **Mixed Strategy**  
  → Select both:
  - Some data with **minimum margin**
  - Some data with **maximum margin**

## ❓ Query Strategies for Active Learning

### 1. Uncertainty Sampling
Select data points for which the current model is **least certain**.

➡️ Current Model = Model trained on current labelled data

---

### 2. Query by Committee
- Train multiple models (committee)
- Select data where **models disagree the most**

---

### 3. Expected Model Change
Select data points that are expected to cause the **maximum change** in model parameters.

---

### 4. Expected Error Reduction
Select data points that will **reduce generalization error the most**

#### 📌 Formula
(Summation i=1→50) (yᵢᵗ - yᵢᵖʳᵉᵈ)²  
−  
(Summation i=1→60) (yᵢᵗ - yᵢᵖʳᵉᵈ)²  

➡️ Maximum gain occurs when adding new informative data (e.g., 50 + 10)

---

### 5. Variance Reduction
Select data points that reduce the **output variance** of the model.

## 🧪 Practical Section
(To be implemented / explored in practice)

### Implementation of Active Learning

#### It shows how **Active Learning** improves a model by repeatedly selecting the most **uncertain data points** for labeling.

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression

#Create Simple dataset(fix parameters properly)
x,y=make_classification(n_samples=100, n_features=2, n_redundant=0, n_informative=2, random_state=0)

#Start with 10 labelled samples
labelled=np.arange(10)
unlabelled=np.arange(10,100)
model=LogisticRegression(solver="liblinear")
for i in range(5):  # 5 active learning queries
    # Train model on labelled data
    model.fit(x[labelled], y[labelled])
    
    # Predict probabilities for unlabelled data
    probs = model.predict_proba(x[unlabelled])[:, 1]
    
    # Select most uncertain sample (closest to 0.5 probability)
    query = np.argmin(np.abs(probs - 0.5))
    
    # Move selected sample from unlabelled → labelled
    labelled = np.append(labelled, unlabelled[query])
    unlabelled = np.delete(unlabelled, query)
    
    print(f"Iteration: {i+1} | Labelled size: {len(labelled)}")
```
