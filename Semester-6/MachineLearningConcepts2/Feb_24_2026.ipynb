# 📊 Association Rule Mining

## 🔍 Finding Patterns Using Association Rules

**Association Rule Mining** is a technique used to discover interesting relationships and patterns hidden in large datasets.

It is a part of **Association Analysis**, where relationships between items are represented as **rules** derived from **frequent itemsets**.

---

# 🛒 Example: Market Basket Analysis

Consider the following transactions in a retail store:

| Transaction ID | Items Purchased |
|---------------|-----------------|
| T1 | Bread, Milk, Egg, Salt |
| T2 | Bread, Milk, Butter |
| T3 | Milk, Egg, Sugar |

### 💡 Insight

If customers frequently buy **Bread and Milk together with Egg**,  
the retailer can:
- Place these items close to each other 🏪
- Create combo offers 🎁
- Offer bundle discounts 💰

---

# 📌 Example of an Association Rule

| Antecedent (LHS) | Consequent (RHS) |
|------------------|------------------|
| {Bread, Milk} | {Egg} |

### Interpretation:
If a customer buys **Bread and Milk**,  
there is a high likelihood they will also buy **Egg**.

This can be written as:

\[
\{Bread, Milk\} ->{Egg\}
\]

# 📖 Common Terminologies

## 1️⃣ Itemset

An **Itemset** is a group of one or more items that appear together in a transaction.

## 📌 Understanding Itemsets

### 🔹 Example of an Itemset

For example:

\[
\{Milk, Egg, Butter\}
\]

This is called a **3-itemset** because it contains **3 items**.

---

## 🔢 Mathematical Representation

Let:

- \( I = \{ i_1, i_2, i_3, \dots, i_m \} \)  
  → be the set of all possible items  

- \( T = \{ t_1, t_2, t_3, \dots, t_n \} \)  
  → be the set of all transactions  

Each transaction \( t_k \subseteq I \),  
meaning every transaction contains a **subset of items** from the complete item set \( I \).

### ✅ Definition

A **k-itemset** is an itemset that contains exactly **k items**.

---

## 🧾 Example Transaction Dataset

| Transaction Number | Purchased Items |
|--------------------|------------------|
| 1 | {Bread, Milk, Egg, Butter, Salt, Apple} |
| 2 | {Bread, Milk, Egg, Apple} |
| 3 | {Bread, Milk, Butter, Apple} |
| 4 | {Milk, Egg, Butter, Apple} |
| 5 | {Bread, Egg, Salt} |
| 6 | {Bread, Milk, Egg, Apple} |

---

## 🎯 Key Concept

- Items → Elements of set \( I \)  
- Transactions → Subsets of \( I \)  
- Itemsets → Groups of items appearing together  
- k-itemset → Contains exactly **k items**
