# üìå K-Means Clustering  
### A **Centroid-Based Clustering Technique**

**K-Means Clustering** is one of the most widely used **unsupervised learning algorithms** for partitioning data into distinct groups.

---

## üîπ Key Concepts

- **Type:** *Centroid-Based Clustering Algorithm*  
- **Centroid:** Each cluster is represented by the **mean of all data points** belonging to that cluster  
- **Requirement:** The **number of clusters (K)** must be predefined  

---

## ‚ùì Why Do We Need K-Means?

- To **automatically partition large datasets** into meaningful groups  
- To **simplify complex data** by reducing it to representative centroids  
- To identify **underlying structure** in **unlabelled data**

---

## üéØ Purpose of K-Means Clustering

- To group **similar data points** based on a **distance metric**  
- To **support decision-making** by analyzing cluster patterns  

---

## üßÆ K-Means Algorithm Steps

1. **Select** `K` initial centroids randomly from the dataset.  
2. **Assign** each data point to the **nearest centroid** using a distance metric.  
3. **Recompute** centroids as the **mean of all data points** in each cluster.  
4. **Reassign** each data point to the **nearest updated centroid**.  
5. **Repeat Steps 3 and 4** until centroids **do not change significantly** (*convergence*).  
6. **Output** the final clusters and their centroids.

---

### üîÅ Iterative Nature of K-Means

> **Note:** Steps **2 to 5** form the **iterative loop** of the K-Means algorithm.

---

## üìê Distance Calculation & Centroid Update in K-Means

### Step 1: Initialize Centroids
- Initialize `K` centroids **randomly** from the dataset.

---

### Step 2: Compute Distance (Euclidean Distance)

The distance between a data point and a centroid is calculated using:

\[
d\big((x, y), (a, b)\big) = \sqrt{(x - a)^2 + (y - b)^2}
\]

---

### Step 3: Assign Clusters (Iteration 1)
- Assign each data point to the **nearest centroid**.
- This completes **Iteration 1** of clustering.

---

### Step 4: Update Centroids
- Compute new centroids as the **mean of all coordinates** in each cluster.

---

### üìç Centroid Update After Iteration 1

- The **new centroids** are the **mean of the coordinates** of data points in their respective clusters.
- These centroids are used for the **next iteration**.

---

> üîÑ The process continues until the centroids **stabilize** and no longer change significantly

# üßë‚Äçüíª K-Means Clustering: Practical Implementation


```python
import numpy as np
import numpy as np

x = np.random.randint(0, 10, size=(10, 2))
print(x)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[:,0],x[:,1])
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Scatter Plot Before Clustering")
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3,random_state=42)
#model=kmeans.fit(x)
#label=kmeans.fit_predict(x)  Writing this 2 steps does the same as below
labels=kmeans.fit_predict(x)
print(labels)
print(kmeans.cluster_centers_)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[:,0],x[:,1],c=labels)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Scatter Plot Before Clustering")
plt.grid(True)
plt.show()
```
