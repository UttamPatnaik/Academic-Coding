```python
from textblob import TextBlob

# Defining the text samples (completed from the image)
text1 = 'Love never came to me in a silver spoon, so I licked it off the floor.'
text2 = 'I was sad. I am happy. I will be elated.'
text3 = 'I am very happy because I was enjoying my time at home during the holidays.'
text4 = '''There are vast organizations in Europe and Asia 
who are rivals. The stories of Europe's exploitations have reached 
far and wide. Other countries have joined forces with Asia's superior 
technology to level the playing field and change the story.'''
text5 = 'This product is awful. It smells disgusting. Very disappointing.'

# Creating TextBlob objects
blob1 = TextBlob(text1)
blob2 = TextBlob(text2)
blob3 = TextBlob(text3)
blob4 = TextBlob(text4)
blob5 = TextBlob(text5)

# Displaying the type and specific sentiment results
print(f"Type of sentiment: {type(blob1.sentiment)}")
print("-" * 30)

# Printing Polarity and Subjectivity for each
blobs = [blob1, blob2, blob3, blob4, blob5]

for i, b in enumerate(blobs, 1):
    print(f"Blob {i} Sentiment: {b.sentiment}")


```

```python
from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer
text="Today is a beatiful day.Tomorrow looks like bad weather"
blob=TextBlob(text,analyzer=NaiveBayesAnalyzer())
for sentence in blob.sentences:
    sentiment=sentence.sentiment
    print("Sentence",sentence)
    print("Classification",sentiment.classification)
    print("Positive Probability: ",sentiment.p_pos)
    print("Negative Probability: ",sentiment.p_pos)
```

```python
#Infection Pluralization and Singularization
from textblob import Word, TextBlob

word1 = Word('index')
word2 = Word('children')

text1 = "I was sad. I was happy. I will be elated"

# Convert text into words
blob = TextBlob(text1)

# Pluralize words individually
plural_words = [word.pluralize() for word in blob.words]

print("Pluralized sentence:", " ".join(plural_words))
print("Plural of 'index':", word1.pluralize())
print("Plural of 'children':", word2.pluralize())
```

## Stemming and Lemmatization

**Stemming** and **Lemmatization** are normal text processing operations that remove prefixes or suffixes from a word, leaving only the stem or root word.

- The root word derived by **stemming** may or may not be a real word.
- Whereas **lemmatization** always derives a meaningful (real) word.

```python
from textblob import Word
from textblob.wordnet import VERB, NOUN, ADJ, ADV
word=Word("run")
verb_synsets=word.get_synsets(pos=VERB)
print("Verb Synsets: ")
for synset in verb_synsets:
    print(synset.name()," : ",synset.definition())

noun_synsets=word.get_synsets(pos=VERB)
print("Noun Synsets: ")
for synset in noun_synsets:
    print(synset.name()," : ",synset.definition())
```

```python
from textblob import Word
word=Word('happy')
for i in word.synsets:
    for lemma in i.lemmas():
        print(lemma)
```

```python
from textblob import Word
synonyms=set()
word=Word('happy')
for i in word.synsets:
    for lemma in i.lemmas():
        synonyms.add(lemma.name())
print(synonyms)
```

## Stop Words

Stop words are common words in a text that are often removed before analyzing it, because they typically do not provide useful information.

The NLTK library provides lists of stop words for several natural languages.

```python
from nltk.corpus import stopwords
from textblob import TextBlob
stops=stopwords.words("english")
blob=TextBlob("Today is a beautiful day")
print([word for word in blob.words if word not in stops])
print(stops)
```

Ingrams is a continuous sequence of words in a text, the default is trigrams

```python
from textblob import TextBlob
n=10
text="Today is a beautiful day. Tomorrow looks like bad weather"
blob=TextBlob(text)
print(blob.ngrams(n))
```

```python
from textblob import TextBlob
n=10
text="Today is a beautiful day. Tomorrow looks like bad weather"
blob=TextBlob(text)
print(blob.ngrams(n<5))
```
