```python
#Default Constructor
class Student:
    def __init__(self):
        self.name="abc"
        self.age=10
        self.marks=100
    def talk(self):
        print("Hi I am",self.name)
        print("My age is",self.age)
        print("My marks are",self.marks)
s1=Student()
s1.talk()
```

```python
#Parameterized Constructor
class Student:
    def __init__(self,name,age,marks):
        self.name=name
        self.age=age
        self.marks=marks
    def talk(self):
        print("Hi I am",self.name)
        print("My age is",self.age)
        print("My marks are",self.marks)
s1=Student("Chatu",10,100)
s1.talk()
```

```python
class Student:
    def __init__(self,name="Gobi"):
        self.name=name
    def talk(self):
        print("Hi I am",self.name)
s1=Student()
s1.talk()
```

```python
#Instance Variable
class Sample:
    def __init__(self):
        self.x=10
    def modify(self):
        self.x+=1
S1=Sample()
S2=Sample()
print("X in S1: ",S1.x)
print("X in S2: ",S2.x)
S1.modify()
print("X in S1: ",S1.x)
print("X in S2: ",S2.x)
```

```python
#Class Variable
class Sample:
    x=10
    @classmethod
    def modify(cls):
        cls.x+=1
S1=Sample()
S2=Sample()
print("X in S1: ",S1.x)
print("X in S2: ",S2.x)
S1.modify()
print("X in S1: ",S1.x)
print("X in S2: ",S2.x)
```
