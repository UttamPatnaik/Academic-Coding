### 2.Create a custom Python class for managing a bank account with basic functionalities like deposit and withdrawal?

```python
class bankAccount:
    def __init__(self, name, acid, bal):
        self.name = name
        self.acid = acid
        self.bal = bal

    def withdraw(self, amount):
        if amount > self.bal:
            print("Insufficient Balance")
        else:
            self.bal -= amount
            print("Remaining Balance:", self.bal)

    def deposit(self, amount):
        self.bal += amount
        print("Current Balance:", self.bal)

def main():
    bal = bankAccount("Kapil Dev Sharma", 7465, 10000)
    bal.withdraw(5000)
    bal.deposit(5000)

main()
```

### 3. Create a Book class that contains multiple Chapters, where each Chapter has a title and page count. Write code to initialize a Book object with three chapters and display the total page count of the book.

```python
class Chapter:
    def __init__(self, title, pages):
        self.title = title
        self.pages = pages

class Book:
    def __init__(self, title):
        self.title = title
        self.chapters = []

    def add_chapter(self, chapter):
        self.chapters.append(chapter)

    def total_pages(self):
        total = 0
        for chapter in self.chapters:
            total += chapter.pages
        return total

my_book = Book("Python Programming")

chapter1 = Chapter("Introduction", 10)
chapter2 = Chapter("Data Types", 25)
chapter3 = Chapter("Control Flow", 30)

my_book.add_chapter(chapter1)
my_book.add_chapter(chapter2)
my_book.add_chapter(chapter3)

print("Total pages in the book:", my_book.total_pages())
```

### 4. How does Python enforce access control to class attributes, and what is the difference between public, protected, and private attributes?

### Python controls access to class attributes using naming conventions.
#### Public attributes are declared normally and can be accessed from anywhere.
#### Protected attributes start with a single underscore (_) and are intended to be used within the class and its subclasses.
#### Private attributes start with double underscores (__) and use name mangling to restrict direct access outside the class.
### However, Python does not strictly enforce access control; it mainly follows conventions.

### 5. Write a Python program using a Time class to input a given time in 24-hour format and convert it to a 12-hour format with AM/PM. The program should also validate time strings to ensure they are in the correct HH:MM:SS format. Implement a method to check if the time is valid and return an appropriate message.

```python
class Time:
    def __init__(self, time_str):
        self.time_str = time_str

    def is_valid(self):
        parts = self.time_str.split(":")

        if len(parts) != 3:
            return False

        try:
            h = int(parts[0])
            m = int(parts[1])
            s = int(parts[2])

            if 0 <= h < 24 and 0 <= m < 60 and 0 <= s < 60:
                return True
            else:
                return False
        except:
            return False

    def convert_12hr(self):
        if not self.is_valid():
            return "Invalid time format"

        h, m, s = map(int, self.time_str.split(":"))

        period = "AM"

        if h >= 12:
            period = "PM"

        if h == 0:
            h = 12
        elif h > 12:
            h -= 12

        return f"{h:02}:{m:02}:{s:02} {period}"

t = input("Enter time in 24-hour format (HH:MM:SS): ")
time_obj = Time(t)

if time_obj.is_valid():
    print("Valid Time")
    print("12-hour format:", time_obj.convert_12hr())
else:
    print("Invalid Time Format")
```
