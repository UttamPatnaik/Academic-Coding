```python
#1. Create a 2-by-3 array with ones, a 3-by-3 array with zeros and a 2-by-5 array with sevens.
import numpy as np
x=np.ones((2,3),dtype=int)
y=np.zeros((2,3),dtype=int)
z=np.full((2,5),7)
print(x)
print(y)
print(z)
```

```python
#2. Use arange to create a 2-by-2 array containing the numbers 0-3. Use broadcasting to perform each
# of the following operations on the original array:
#(a) Cube every element of the array.
#(b) Add 7 to every element of the array.
#(c) Multiply every element of the array by 2.
import numpy as np
x=np.arange(0,4).reshape(2,2)
print("Original Array \n",x)
print("Cube \n",x**3)
print("Adding 7 to each element \n",x+7)
print("Multiplying 2 to each element \n",x*2)
```

```python
#3. Create a 3-by-3 array containing the even integers from 2 through 18. Create a second 3-by-3
# array containing the integers from 9 down to 1, then multiply the first array by the second.
import numpy as np
x=np.arange(2,19,2).reshape(3,3)
print("arr1 \n",x)
y=np.arange(9,0,-1).reshape(3,3)
print("arr2 \n",y)
z=x*y
print("Resultant Array \n",z)
```

```python
#4. Create a 2D array and swap the first two rows and two columns.
import numpy as np
x=np.array([[1,2],[3,4]])
print("Original Array \n",x)
y=x[[1,0]]
print("Swapping first 2 rows \n",y)
z=x[:,[1,0]]
print("Swapping first 2 columns \n",z)
```

```python
#4. Create a 2D array and swap the first two rows and two columns.
import numpy as np
x=np.array([[1,2],[3,4]])
x1=x.copy()
x2=x.copy()
print("Original Array \n",x)
x1[[0,1]]=x1[[1,0]]
print("Swapping first 2 rows \n",x1)
x1[:,[0,1]]=x1[:,[1,0]]
print("Swapping first 2 columns \n",x1)
```

```python
#5. Create a 2-by-5 array from an argument which is a list of the two five-element
#lists [2, 3, 5, 7, 11] and [13, 17, 19, 23, 29].
import numpy as np
l1=[[2,3,5,7,11],[13,17,19,24,29]]
x=np.array(l1).reshape(2,5)
print(x)
```

```python
#6. Create a 2-by-3 array containing the first six powers of 2 beginning with 2^0.
#Flatten the array first with method flatten, then with ravel. In each case,
#display the result then display the original array to show that it was unmodified.
import numpy as np
a=np.array([2**i for i in range(0,6)]).reshape(2,3)
print("Original array: \n",a)
a1=a.flatten()
print("Flattened array: \n",a1)
a2=a.ravel()
print("Ravelled array: \n",a2)
```
