```python
gb={"Bunu":[97,91,89],"Pantipa":[95,85,88],"Azizi":[92,86,90],"Jinn":[81,98,79]}
gt=0
gc=0
for n,g in gb.items():
    t=sum(g)
    print(f"Average for {n} is: {t/len(g)}")
    gt+=t
    gc+=len(g)
print(f"Class Average: {gt/gc}")
```

```python
text = "Sample text with several words This is more sample text with some different words"
wc = {}

for w in text.split():
    if w in wc:
        wc[w] += 1
    else:
        wc[w] = 1

print(f"{'Word':<12} Count")
for w, c in sorted(wc.items()):
    print(f"{w:<12}{c}")
print("Number of unique words:", len(wc))
```
